# Random Forest

## 1. Bootstrap sampling

```
N = 100
```

```
training_data = list()
for i in range(N):
  X_sp, X_dp, y_sp, y_dp = train_test_split(X_train, y_train, test_size=0.2, random_state= i+6)
  training_data.append({'X' : X_sp, 'y' : y_sp})
```

## 2. Train separately

```
classifier_forest = list()
for i in range(N):
  tree = DecisionTreeClassifier(max_features='sqrt', random_state=i+2)    #ออกแบบให้ต้นไม้มีลักษณะพิเศษ
  tree.fit(training_data[i]['X'], training_data[i]['y'])
  classifier_forest.append(tree)
```

## 3. To predict, aggregating : เก็บผลการทำนายไว้ใน list

```
prediction = list()
for i in range(N):
  pred = classifier_forest[i].predict(X_test)
  # print pred
  prediction.append(pred)
```

```
prediction
```

```
from scipy import stats
import numpy as np
```

```
m = stats.mode(np.array(prediction))
m
```

## 4. Evalution

```
from sklearn.metrics import accuracy_score
```

```
accuracy_score(y_test, m.mode.squeeze())      # มีความแม่นยำกว่า
```
